```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.boxplot(df['Age'])
```

```python
sns.boxplot(df['Fare'])
```

```python
sns.boxplot(df['SibSp'])
```

```python
sns.catplot(x= 'Pclass', y = 'Age', data=df, kind = 'box')
```

```python
sns.boxplot(x='Sex', y='Age', hue='Survived', data=df)
```

```python
sns.violinplot(x='Sex', y='Age', hue='Survived', data=df)
```

```python
sns.swarmplot(x='Sex', y='Age', hue='Survived', data=df)
```

```python
sns.stripplot(x='Sex', y='Age', hue='Survived', data=df)
```
